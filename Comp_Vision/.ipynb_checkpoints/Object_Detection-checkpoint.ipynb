# 3. Object Detection

##### Training

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 1.Resize the training images to 128 x 128

# load images
test = cv2.cvtColor(cv2.imread("test.jpg"), cv2.COLOR_BGR2RGB)
car_path = ['car1.jpg', 'car2.jpg', 'car3.jpg', 'car4.jpg', 'car5.jpg', 'car6.jpg', 'car7.jpg', 'car8.jpg', 'car9.jpg',
           'car10.jpg', 'car11.jpg', 'car12.jpg', 'car13.jpg', 'car14.jpg']

cars = []
for i in range(0, len(car_path)):
    # Read each car image and make into grayscale
    car = cv2.imread(car_path[i], cv2.IMREAD_GRAYSCALE)
    
    # Resize image
    car = cv2.resize(car, (128, 128))
    
    # Append each image to the list
    cars.append(car)

```

```python
# 2. Compute HoG features using a cell size of 4x4 pixels, block size of 2x2 cells and 9 orientation bins

# h x w in pixels
cell_size = (4,4) 
# h x w in cells
block_size = (2,2) 
# number of orientation bins
nbins  = 9

def HoG_features(list):
    # List of images as arguments
    cells = []
    # Empty HoG list for delivery
    hog_feat = []
    
    # Create HoG object
    for i in range(0, len(list)):
        hog = cv2.HOGDescriptor(_winSize = (list[i].shape[1]//cell_size[1]*cell_size[1], 
                                            list[i].shape[0]//cell_size[0]*cell_size[0]),
                                _blockSize = (block_size[1]*cell_size[1], 
                                              block_size[0]*cell_size[0]),
                                _blockStride = (cell_size[1], cell_size[0]),
                                _cellSize = (cell_size[1], cell_size[0]),
                                _nbins = nbins)
        
        n_cells = (list[i].shape[0]//cell_size[0], list[i].shape[1]//cell_size[1])
        cells.append(n_cells)
        
        # Compute HoG features
        hog_feats = hog.compute(list[i]) \
                               .reshape(cells[i][1] - block_size[1] + 1,
                                       cells[i][0] - block_size[0] + 1,
                                       block_size[1], block_size[0], nbins) \
                                       .transpose((1, 0 , 3, 2, 4))
        
        hog_feat.append(hog_feats)
    
    # return HoG features for each image in the list of HoG features
    return hog_feat

# Obtain list of HoG features
hog_features_list = HoG_features(cars)
```

```python
np.shape(hog_features_list)
```

```python
# 3. Calculate and store the mean feature map across training images

# Calculate the mean of features for training images
sum = 0
# loop through to compute sum
for i in range(0, len(hog_features_list)):
    sum += hog_features_list[i]
# calculate average
mean = sum/len(hog_features_list)
```

```python
# 4. Repeat above steps for training images flipped around vertical axis

cars_copy = cars.copy()

# Flip each of the training images around vertical axis
for i in range(0, len(cars_copy)):
    cars_copy[i] = cv2.flip(cars_copy[i], 0)

# Compute HoG features
hog_features_flipped = HoG_features(cars_copy)

# Calculate the mean
sum_flipped = 0
for i in range(0, len(hog_features_flipped)):
    sum_flipped += hog_features_flipped[i]

mean_flipped = sum_flipped/len(hog_features_flipped)
```
