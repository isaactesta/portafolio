In this project, the relationship between emotional states and EEG signals was explored. The data (DEAP: A dataset for Emotional Analysis using Physiological Signals) consists of the subjective response of patients that watched a youtube video with a strong emotional components, while their EEG signals were being recorded.


### Imports

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import signal
from scipy.integrate import simps
from scipy.stats.stats import pearsonr
```

### Open Data Files

```python
x = pickle.load(open(rb'C:\Users\isaac\Desktop\DEAP\data_preprocessed_python\data_preprocessed_python\s01.dat', 'rb'),encoding='latin1')
participants_ratings = pd.read_excel(r'C:\Users\isaac\Desktop\DEAP\metadata_xls\participant_ratings.xls')
```

Seprating Dictionary to arrays

```python
labels = x.get('labels')
data = x.get('data')
```

Data Plotting

```python
sns.set()
plt.figure(0)
plt.plot(data[0][0])
plt.show()
```

Defining Frequency bands of the brain:

Theta: 3 - 7 Hz  
Alpha: 8 - 13 Hz  
Beta: 14 - 29 Hz  
Gamma: 30 - 47 Hz

```python
# Theta
low_t, high_t = 3, 7
# Alpha
low_a, high_a = 8, 13
# Beta
low_b, high_b = 14, 29
# Gamma
low_g, high_g = 30, 47
```

Using Welch's method to plot obtain and plot the power spectra

```python
fs = 128
win = 4*128

f,pxx = signal.welch(data[0][0], fs, nperseg=win)


sns.set()
plt.figure(1)
plt.plot(f,pxx)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density (V^2/Hz)')
plt.ylim([0, pxx.max()*1.1])
plt.title("Example of Welch's Periodogram of EEG signal")
sns.despine()
```

Deifinig bands in periodogram

```python
idx_theta = np.logical_and(f >= low_t, f <= high_t+1)
idx_alpha = np.logical_and(f >= low_a, f <=high_a+1)
idx_beta = np.logical_and(f >= low_b, f <=high_b+1)
idx_gamma = np.logical_and(f >= low_g, f <=high_g+1)


sns.set()
plt.figure(2)
plt.plot(f,pxx)
plt.fill_between(f,pxx, where = idx_theta, color='skyblue')
plt.fill_between(f,pxx, where = idx_alpha, color='lightgreen')
plt.fill_between(f,pxx, where = idx_beta, color='red')
plt.fill_between(f,pxx, where = idx_gamma, color='burlywood')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density ($V^2$/Hz)')
plt.ylim([0, pxx.max()*1.1])
plt.xlim([0, 50])
plt.title("Welch's Periodogram of EEG signal with frequency bands defined")
plt.legend(['PSD','Theta band','Alpha band','Beta band','Gamma band'], frameon=False)
sns.despine()
```

Channel Selection:  

Emotional response has been shown to be Frontal and Assymetric so channels were chosen accordingly in order to compare the correlations.  
Channels choosen: FC5(4), T7(7), CP1(9), FC6(21), CZ(24), T8(26), and CP6(27)


```python
channels = [4,7,9,21,23,25,26]
s = (7,257)
F = np.zeros(s)
pxx_s = np.zeros(s)

freqs_res = f[1]-f[0]

theta_power = np.zeros(7)
alpha_power = np.zeros(7)
beta_power = np.zeros(7)
gamma_power = np.zeros(7)


for i in range(len(channels)):
    c = channels[i]
    F[i], pxx_s[i] = signal.welch(data[2][c],fs,nperseg=win)
    idx_theta = np.logical_and(F[i] >= low_t, F[i] <= high_t+1)
    idx_alpha = np.logical_and(F[i] >= low_a, F[i] <=high_a+1)
    idx_beta = np.logical_and(F[i] >= low_b, F[i] <=high_b+1)
    idx_gamma = np.logical_and(F[i] >= low_g, F[i] <=high_g+1)
    
    theta_power[i] = simps(pxx_s[i,idx_theta], dx = freqs_res)
    alpha_power[i] = simps(pxx_s[i,idx_alpha], dx = freqs_res)
    beta_power[i] = simps(pxx_s[i,idx_beta], dx = freqs_res)
    gamma_power[i] = simps(pxx_s[i,idx_gamma], dx = freqs_res)    

    
```

```python
t = (40,7,257)
F_2 = np.zeros(t)
pxx_2 = np.zeros(t)

freqs_res = f[1]-f[0]

theta_power_2 = np.zeros((40,7))
alpha_power_2 = np.zeros((40,7))
beta_power_2 = np.zeros((40,7))
gamma_power_2 = np.zeros((40,7))


for j in range(0,39):
    for i in range(len(channels)):
        c = channels[i]
        F_2[j,i], pxx_2[j,i] = signal.welch(data[j][c],fs,nperseg=win)
        idx_theta = np.logical_and(F_2[j,i] >= low_t, F_2[j,i] <= high_t+1)
        idx_alpha = np.logical_and(F_2[j,i] >= low_a, F_2[j,i] <=high_a+1)
        idx_beta = np.logical_and(F_2[j,i] >= low_b, F_2[j,i] <=high_b+1)
        idx_gamma = np.logical_and(F_2[j,i] >= low_g, F_2[j,i] <=high_g+1)
    
        theta_power_2[j,i] = simps(pxx_2[j,i,idx_theta], dx = freqs_res)
        alpha_power_2[j,i] = simps(pxx_2[j,i,idx_alpha], dx = freqs_res)
        beta_power_2[j,i] = simps(pxx_2[j,i,idx_beta], dx = freqs_res)
        gamma_power_2[j,i] = simps(pxx_2[j,i,idx_gamma], dx = freqs_res)   
        
```

```python
t = (40,40,257)
F_2 = np.zeros(t)
pxx_2 = np.zeros(t)

freqs_res = f[1]-f[0]

theta_power_2 = np.zeros((40,40))
alpha_power_2 = np.zeros((40,40))
beta_power_2 = np.zeros((40,40))
gamma_power_2 = np.zeros((40,40))


for j in range(0,40):
    for i in range(0,40):       
        
        F_2[j,i], pxx_2[j,i] = signal.welch(data[j][i],fs,nperseg=win)
        idx_theta = np.logical_and(F_2[j,i] >= low_t, F_2[j,i] <= high_t+1)
        idx_alpha = np.logical_and(F_2[j,i] >= low_a, F_2[j,i] <=high_a+1)
        idx_beta = np.logical_and(F_2[j,i] >= low_b, F_2[j,i] <=high_b+1)
        idx_gamma = np.logical_and(F_2[j,i] >= low_g, F_2[j,i] <=high_g+1)
    
        theta_power_2[j,i] = simps(pxx_2[j,i,idx_theta], dx = freqs_res)
        alpha_power_2[j,i] = simps(pxx_2[j,i,idx_alpha], dx = freqs_res)
        beta_power_2[j,i] = simps(pxx_2[j,i,idx_beta], dx = freqs_res)
        gamma_power_2[j,i] = simps(pxx_2[j,i,idx_gamma], dx = freqs_res)   
```

Arraging the results obtained from the Russell scale:

```python
valence = labels[:,0]
arousal = labels[:,1]
dominance = labels[:,2]
liking = labels[:,3]
```

Calculating Pearson Correlation coefficients

```python
corr_t = np.zeros((31,4))
p_val_t = np.zeros((31,4))

corr_a = np.zeros((31,4))
p_val_a = np.zeros((31,4))

corr_b = np.zeros((31,4))
p_val_b = np.zeros((31,4))

corr_g = np.zeros((31,4))
p_val_g = np.zeros((31,4))

for i in range(0,31): 
    corr_t[i,0],p_val_t[i,0] = pearsonr(valence, theta_power_2[:,i])
    corr_a[i,0],p_val_a[i,0] = pearsonr(valence, alpha_power_2[:,i])
    corr_b[i,0],p_val_b[i,0] = pearsonr(valence, beta_power_2[:,i])
    corr_g[i,0],p_val_g[i,0] = pearsonr(valence, gamma_power_2[:,i])
    
    corr_t[i,1],p_val_t[i,1] = pearsonr(arousal, theta_power_2[:,i])
    corr_a[i,1],p_val_a[i,1] = pearsonr(arousal, alpha_power_2[:,i])
    corr_b[i,1],p_val_b[i,1] = pearsonr(arousal, beta_power_2[:,i])
    corr_g[i,1],p_val_g[i,1] = pearsonr(arousal, gamma_power_2[:,i])
    
    corr_t[i,2],p_val_t[i,2] = pearsonr(dominance, theta_power_2[:,i])
    corr_a[i,2],p_val_a[i,2] = pearsonr(dominance, alpha_power_2[:,i])
    corr_b[i,2],p_val_b[i,2] = pearsonr(dominance, beta_power_2[:,i])
    corr_g[i,2],p_val_g[i,2] = pearsonr(dominance, gamma_power_2[:,i])
    
    corr_t[i,3],p_val_t[i,3] = pearsonr(liking, theta_power_2[:,i])
    corr_a[i,3],p_val_a[i,3] = pearsonr(liking, alpha_power_2[:,i])
    corr_b[i,3],p_val_b[i,3] = pearsonr(liking, beta_power_2[:,i])
    corr_g[i,3],p_val_g[i,3] = pearsonr(liking, gamma_power_2[:,i])
```

Plots with bands correlations and emotional response

```python
f,axarr = plt.subplots(2,2)
plt.subplots_adjust(wspace = 0.5,hspace=0.5)

axarr[0,0].hist(corr_t[:,0], color='skyblue')
axarr[0,0].set_title('Valence')
axarr[1,0].hist(corr_t[:,1], color='skyblue')
axarr[1,0].set_title('Arousal')
axarr[0,1].hist(corr_t[:,2], color='skyblue')
axarr[0,1].set_title('Dominance')
axarr[1,1].hist(corr_t[:,3], color='skyblue')
axarr[1,1].set_title('Liking')
f.suptitle('Theta Band correlations')
plt.savefig('thetaband.png')
plt.show()

f,axarr = plt.subplots(2,2)
plt.subplots_adjust(wspace = 0.5,hspace=0.5)
axarr[0,0].hist(corr_a[:,0], color='lightgreen')
axarr[0,0].set_title('Valence')
axarr[1,0].hist(corr_a[:,1], color='lightgreen')
axarr[1,0].set_title('Arousal')
axarr[0,1].hist(corr_a[:,2], color='lightgreen')
axarr[0,1].set_title('Dominance')
axarr[1,1].hist(corr_a[:,3], color='lightgreen')
axarr[1,1].set_title('Liking')
f.suptitle('Alpha Band correlations')
plt.savefig('alphaband.png')
plt.show()

f,axarr = plt.subplots(2,2)
plt.subplots_adjust(wspace = 0.5,hspace=0.5)
axarr[0,0].hist(corr_b[:,0], color='red')
axarr[0,0].set_title('Valence')
axarr[1,0].hist(corr_b[:,1], color='red')
axarr[1,0].set_title('Arousal')
axarr[0,1].hist(corr_b[:,2], color='red')
axarr[0,1].set_title('Dominance')
axarr[1,1].hist(corr_b[:,3], color='red')
axarr[1,1].set_title('Liking')
f.suptitle('Beta Band correlations')
plt.savefig('betaband.png')
plt.show()

f,axarr = plt.subplots(2,2)
plt.subplots_adjust(wspace = 0.5,hspace=0.5)
axarr[0,0].hist(corr_g[:,0], color='burlywood')
axarr[0,0].set_title('Valence')
axarr[1,0].hist(corr_g[:,1], color='burlywood')
axarr[1,0].set_title('Arousal')
axarr[0,1].hist(corr_g[:,2], color='burlywood')
axarr[0,1].set_title('Dominance')
axarr[1,1].hist(corr_g[:,3], color='burlywood')
axarr[1,1].set_title('Liking')
plt.savefig('gammaband.png')
f.suptitle('Gamma Band correlations')
plt.show()
```
